# Traffic Prediction

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# we need to first load the dataset
dataset_path = r"/content/Traffic dataset.csv"
dataframe = pd.read_csv(dataset_path)

```

```python
dataframe.head()
```

```python
dataframe.info()
```

```python

# Descriptive statistics
dataframe.describe()
```

```python
# Visualization of numerical features
sns.histplot(dataframe['CarCount'], bins=20, kde=True)
plt.title('Distribution of CarCount')
plt.show()
```

```python
# Unique values for categorical features
for column in dataframe.select_dtypes(include='object').columns:
    print(f'{column}: {dataframe[column].unique()}')
    print(f'Counts:\n{dataframe[column].value_counts()}\n')

```

```python
# Visualization of categorical features
sns.countplot(x="Day of the week", data=dataframe)
plt.title('Distribution of DayOfWeek')
plt.show()
```

```python
# Identify outliers
sns.boxplot(x=dataframe['CarCount'])
plt.title('Boxplot of CarCount to Identify Outliers')
plt.show()
```

```python

# Visualize outliers using a boxplot
sns.boxplot(x=dataframe['BikeCount'])
plt.title('Boxplot of BikeCount to Identify Outliers')
plt.show()
```
